### distance between queries




```python
distances = []
for session in sessions:
    distance = []
    currentQuery = ""
    for query in session:
        if query[5] == 'Q':
            if currentQuery =="":
                currentQuery = query[1]
                distance.append(0)
            else:
                distance.append(levenshtein_distance(currentQuery,query[1]))
#                 if(levenshtein_distance(currentQuery,query[1]) == 0):
#                     print('here')
#                     print(session)
                currentQuery = query[1]
        else:
            distance.append(-1)
    distances.append(distance)
    distance = []
```

```python
distancesWord = []
for session in sessions:
    distance = []
    currentQuery = ""
    for query in session:
        if query[5] == 'Q':
            if currentQuery =="":
                currentQuery = query[1]
                distance.append(0)
            else:
                distance.append(levenshtein_distance(currentQuery,query[1]))
#                 if(levenshtein_distance(currentQuery,query[1]) == 0):
#                     print('here')
#                     print(session)
                currentQuery = query[1]
        else:
            distance.append(-1)
    distancesWord.append(distance)
    distance = []
```

```python
timeQueries = []
for session in sessions:
    distance = []
    currentTime = -1
    for query in session:
        if query[5] == 'Q':
            if currentTime ==- 1:
                currentTime = query[2]
                distance.append(0.0)
            else:
                distance.append(float(query[2])-float(currentTime))
                currentTime = query[2]
        else:
            distance.append(-1)
    timeQueries.append(distance)
    distance = []
```

### repeated queries

```python
uniqueQueries = []
allSameQueries = []
for session in sessions:
    repeat = []
    currentQuery = ""
    count = 0
    isThere = False
    queries = []
    unique = []
    same = []
    for query in session:
        if query[5] == 'Q':
            queries.append(query[1])
#             for query2 in session:
#                 if query2[5] == 'Q':
#                     if query == query2:
#                         repeat.append(0)
#                         break
#                     if query[1] == query2[1]:
# #                         print('here')
# #                         print(query)
# #                         print(query2)
#                         repeat.append(1)
#                         break

    numUniqQueries = len(set(queries))
    for query in session:
        unique.append(numUniqQueries)
        if (numUniqQueries == 1): 
            if len(queries) >1:
                same.append(1)
            else:
                same.append(-1)
        else:
            same.append(0)
            
    uniqueQueries.append(unique)
    allSameQueries.append(same)
    repeat = []
```

```python
repeatQueries = []
for session in sessions:
    repeat = []
    currentQuery = ""
    count = 0
    isThere = False
    for query in session:
        if query[5] == 'Q':
            for query2 in session:
                if query2[5] == 'Q':
                    if query == query2:
                        repeat.append(0)
                        break
                    if query[1] == query2[1]:
#                         print('here')
#                         print(query)
#                         print(query2)
                        repeat.append(1)
                        break

        else:
            repeat.append(0)
    repeatQueries.append(repeat)
    repeat = []
```

### click on the same result twice

```python
uniqueClicks = []
allSameClicks = []
for session in sessions:
    repeat = []
    currentQuery = ""
    count = 0
    isThere = False
    clicks = []
    unique = []
    same = []
    for query in session:
        if query[5] == 'C':
            clicks.append(query[4])
#             for query2 in session:
#                 if query2[5] == 'Q':
#                     if query == query2:
#                         repeat.append(0)
#                         break
#                     if query[1] == query2[1]:
# #                         print('here')
# #                         print(query)
# #                         print(query2)
#                         repeat.append(1)
#                         break

    numUniqClicks = len(set(clicks))
    for query in session:
        unique.append(numUniqClicks)
        if (numUniqClicks == 1): 
            if len(clicks) >1:
                same.append(1)
            else:
                same.append(-1)
        else:
            same.append(0)
            
    uniqueClicks.append(unique)
    allSameClicks.append(same)
    repeat = []
```

```python
repeatClicks = []
for session in sessions:
    repeat = []
    currentQuery = ""
    count = 0
    isThere = False
    for query in session:
        if query[5] == 'C':
            for query2 in session:
                if query2[5] == 'C':
                    if query == query2:
                        repeat.append(0)
                        break
                    if query[4] == query2[4]:
#                         print('here')
#                         print(query)
#                         print(query2)
                        repeat.append(1)
                        break

        else:
            repeat.append(0)
    repeatClicks.append(repeat)
    repeat = []
```

```python
timeClicks = []
for session in sessions:
    distance = []
    currentTime = -1
    check = False
    for query in session:
        if query[5] == 'C':
            if currentTime ==- 1:
                currentTime = query[2]
                distance.append(0.0)
            elif check == True:
                distance.append(-1)
            else:
                distance.append(float(query[2])-float(currentTime))
                currentTime = query[2]
                check = True
        else:
            distance.append(-1)
            check = False
    timeClicks.append(distance)
    distance = []
```

```python
clickDistance = []
for session in sessions:
    distance = []
    for query in session:
        if query[5] == 'C':
            distance.append(levenshtein_distance(query[4], query[1]))
        else:
            distance.append(-1)
    clickDistance.append(distance)
```

```python
numQ = pd.DataFrame(data = allSessions[allSessions['type']=='Q'].groupby('sID')['type'].count())
numQ = numQ.rename(columns = {'type': 'numQueries'})
numC = pd.DataFrame(data = allSessions[allSessions['type']=='C'].groupby('sID')['type'].count())
numC = numC.rename(columns = {'type': 'numClicks'})
allSessionsC = allSessions[allSessions['type']=='C']
allSessionsC = allSessionsC.astype({'click': 'int32'})
avgC = pd.DataFrame(data = allSessionsC.groupby('sID').mean()['click'])
avgC = avgC.rename(columns = {'click': 'meanClickPosition'})
```

```python
allSessions = allSessions.join(numQ, on = 'sID')
allSessions = allSessions.join(numC, on = 'sID')
allSessions = allSessions.join(avgC, on = 'sID')
allSessions['numClicksPerQuery'] = allSessions['numQueries']/allSessions['numClicks']
```

```python
sessionsClick = allSessions.loc[allSessions['timeClicks'] != -1]
allSessions['timeClicks'] = sessionsClick.groupby('sID')['timeClicks'].mean().fillna(0)
```
